```python
# what I have learned: use hdbscan on the raw points, not the cosine distances

import sqlite3
import pandas as pd
from timeit import default_timer as timer
import numpy as np
import math

db_name = 'abstracts_subsample10'
subsampled_similarity_db = 'data/' + db_name + '.db' # ~250,000 rows

# connect to the database
db = sqlite3.connect(subsampled_similarity_db)
print('database connected')

start = timer()

################### initialize the similarity matrix
# since there are only ~250000 entries, just read edges into a numpy matrix

c = db.cursor()
# get number of rows (vertices)
c.execute('SELECT COUNT(*) from {}'.format(db_name)) # this is small so it's ok to just count
N_edges = c.fetchone()[0]
print(N_edges)
N_vertices = int(math.sqrt(N_edges)) # |E| = |V|^2 and there are no missing edges
W_distance = np.zeros((N_vertices, N_vertices)) # [source, target]
weights_list = []

#c.execute('SELECT * FROM {} LIMIT 10'.format(db_name))
df = pd.read_sql_query('SELECT * FROM {};'.format(db_name), db)
print(df.head())

# translate real indexes to subsampled indexes


for row_idx,row in df.iterrows():
    pre_idx = row['newSourceID']
    post_idx = row['newTargetID']
    weight = row['weight']
    W_distance[pre_idx][post_idx] = 1 - weight # express as distances, not similarity
    weights_list.append(1 / weight)

end = timer()
duration = end - start;
print("read into dataframe: ")
print(duration)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.hist(weights_list)
plt.title('histogram of weights')
plt.show()

plt.figure()
plt.imshow(W_distance)
plt.colorbar()
plt.title('similarity matrix among subsampled vertices')
plt.show()

plt.figure()
im = plt.imshow(W_distance, cmap='hot')
plt.colorbar(im, orientation='horizontal')
plt.title('similarity matrix among subsampled vertices')
plt.show()
```

```python
# try out hdbscan
import hdbscan
#import seaborn as sns

# input is samples x features

clusterer = hdbscan.HDBSCAN(metric='precomputed',cluster_selection_method='leaf',min_samples=1,min_cluster_size=2)
clusterer.fit(W_distance)

print(clusterer.labels_.max())
cluster_labels = clusterer.labels_.tolist()

plt.figure()
plt.plot(clusterer.labels_)
plt.title('cluster labels')
plt.show()

plt.figure()
clusterer.condensed_tree_.plot() #select_clusters=True,
                               # selection_palette=sns.color_palette('deep', 8))
plt.show()

cluster_sizes = dict((x,cluster_labels.count(x)) for x in set(cluster_labels))
print(cluster_sizes)
```
