```python

import sys  
from PyQt4.QtGui import *  
from PyQt4.QtCore import *  
from PyQt4.QtWebKit import *  


class Render(QWebPage):  
  def __init__(self, url):  
    self.app = QApplication(sys.argv)  
    QWebPage.__init__(self)  
    self.loadFinished.connect(self._loadFinished)  
    self.mainFrame().load(QUrl(url))  
    self.app.exec_()  
  
  def _loadFinished(self, result):  
    self.frame = self.mainFrame()  
    self.app.quit()  
    
def get_html(link):        
        q = Request(link)
        q.add_header('User-Agent', 'Mozilla/5.0')
        print(str(q))
        r = urlopen(q).read()
        print(r)
        return r
```

```python
from bs4 import BeautifulSoup
from urllib2 import Request, urlopen

# selector page
# e.g. http://journals.plos.org/ploscompbiol/volume#2018
edition_selector_page = "http://journals.plos.org/ploscompbiol/volume#2017"
list_of_editions = []
print repr(edition_selector_page)
print edition_selector_page

def get_html(link):        
    q = Request(link)
    q.add_header('User-Agent', 'Mozilla/5.0')
    r = urlopen(q).read()
    return r

r = get_html(edition_selector_page)

soup = BeautifulSoup(r,"lxml")

year = 2019 # hacky counter for keeping track of year
links = soup.find_all('a',href=True)  # this returns all the articles, not just the articles for #2017
for link in links:
    
    if len(link.select('img')) > 0: # check for null
        altText = link.select('img')[0]['alt'] 
        if "Journal Cover" in altText: # this is a link we want
            postfix = link['href']
            edition_link = "http://journals.plos.org" + postfix # turn relative link into full path
            month = altText.split("Journal Cover")[0] # month
            month = month.split(' ')[0]
            if "January" in month:
                year = year-1
            print(month, year)
            list_of_editions.append({"url":edition_link,"year":year,"month":month})
            
print(len(list_of_editions))

# todo write list of editions to a json file
```

```python
import json

save_path = "data/ploscompbio_editions.json"

with open(save_path, 'w') as outfile:  
    json.dump(list_of_editions, outfile)
    
print('saved to json file')
```
